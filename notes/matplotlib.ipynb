```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
xs = np.linspace(0, 2*np.pi, 400)
ys = np.sin(xs**2)

xc = np.linspace(0, 2*np.pi, 600)
yc = np.cos(xc**2)
```

```python
# Usual way to create plot
fig, ax = plt.subplots(2, 2, figsize=(16,8))
ax[0,0].plot(xs, ys)
ax[1,1].plot(xs, ys)
ax[0,1].plot(xc, yc)
ax[1,0].plot(xc, yc)
fig.suptitle('Basic plotting using matplotlib')
plt.show()
```

## Figure
A `Figure` is your main (imaginary) canvas. This is where plotting/drawing is done. A figure has a size defined for it at the time of creation. **This is the central object with which you will always be interacting.**

```python
# To define a figure
fig = mpl.figure.Figure(figsize=(10,10))

# OR
fig = plt.figure(figsize=(10,10))
```

## Axes
A Figure by itself does not have any place to plot. You need to attach/add `Axes` to it to do any kind of plotting. You can put as many Axes objects as you want inside of any Figure you have created.

An Axes has:
1. Space (like a blank page) where you can plot.
2. Parent `Figure`.
3. Propoerties stating where it will be placed inside it's parent `Figure`.
4. Methods to draw/plot different kinds of data.

```python
# To create an axis
ax1 = mpl.axes.Axes(fig=fig, rect=[0, 0, 0.8, 0.8], facecolor='red')

# OR
ax1 = plt.Axes(fig=fig, rect=[0, 0, 0.8, 0.8], facecolor='red') #[left_position, bottom_position, height, width] represented as percentages
```

## Using matplotlib's primitves

```python
# Create figure (canvas)
fig = plt.figure(figsize=(6,6))

# Create Axes (notes these are not yet connected to fig)
ax1 = plt.Axes(fig=fig, rect=[0.3, 0.3, 0.4, 0.4], facecolor='red')
ax2 = plt.Axes(fig=fig, rect=[0, 0, 1, 1], facecolor='blue')
```

```python
# Add Axes to the Figure
fig.add_axes(ax2)
fig.add_axes(ax1)
```

```python
fig
```

```python
# Axes can also be removed
fig.delaxes(ax1)
```
