Now the problem is that whenever you see \*\*kwargs you know that you will not be able to know what all this function takes. I do not know what they are and I have to loop at the core libraries to find the arguments. Now to analyze how deep this problem is we can write a function to check the number of occurrences.

Why this problem exists? If we see the libraries that provide us with these keyword arguments we see that most of these libraries provide us with a lot of conveniences. When you are writing a core library that fulfills a fundamental task, you know what arguments those functions take, but when you are writing a convenience on top to wrap those pieces of functionality you do not want to create a dependency between the convenience layer and the actual layer that actual code that it is wrapping. If you were to create that dependency, then whenever you change the core you have to change the dependency as well which is a lot of burden and error prone process.

How can we now improve the documentation with these kwargs.

Two possible solutions.

Static Analysis. Even in C it is almost impossible. Static Analysis in python is extremely hard. `eval` in python essentially makes it impossible. We can create a nodeVisitor object to view all functions. But we will quickly loose all the bindings. There is unbounded complexity.

Dynamic. It is still very very difficult. Impossible in many cases.
